```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stock_data = pd.read_csv('results.csv')
```

```python
stock_data
```

# Hourly High Price Trend

```python
company_arr = stock_data['Company'].unique()
for company in company_arr:
    plot_df = stock_data[stock_data['Company'] == company]
    plt.plot(plot_df.Hour, plot_df.High)
    plt.title(f"Hourly High Price Trend of '{company}'")
    plt.xlabel("Hour")
    plt.ylabel("Hourly High Price")
    plt.show()
```

# Highest Price on Nov 30th, 2021

```python
highest_df = stock_data.groupby('Company').max()['High'].to_frame()
```

```python
plt.bar(highest_df.index, highest_df.High)
plt.title("Highest Price on Nov 30th, 2021")
plt.xlabel("Stock")
plt.ylabel("Price")
plt.show()
```
